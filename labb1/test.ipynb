```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
# Replace this with your actual data
data = pd.DataFrame({
    'Feature1': np.random.randn(100) * 10 + 50,
    'Feature2': np.random.randn(100) * 5 + 30,
    'Feature3': np.random.randn(100) * 15 + 100
})

# Introduce some outliers
data.loc[::10, 'Feature1'] = np.random.uniform(100, 150, size=10)

# Step 1: Visualize the scatter matrix
sns.pairplot(data)
plt.show()

# Step 2: Define criteria for outlier detection using IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Define the outlier criteria
outlier_condition = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))

# Step 3: Remove outliers
# Create a boolean mask for rows without outliers
mask = ~outlier_condition.any(axis=1)
cleaned_data = data[mask]

# Optional: Visualize the cleaned data
sns.pairplot(cleaned_data)
plt.show()

# Display the original and cleaned data shapes
print("Original data shape:", data.shape)
print("Cleaned data shape:", cleaned_data.shape)
```
